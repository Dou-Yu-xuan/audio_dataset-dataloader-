```python
import pandas as pd
import numpy as np
```

```python
data_path = '/workspace/datasets/callCenterDataset'
csv_local_path = "/callCenterDataset/manual-labels - operator - 12-07-18_23_50.csv"
```

```python
df = pd.read_csv(data_path+csv_local_path)
n_samples = df.shape[0]

df['file_name']=df['file_name'].apply(lambda x: '{}/{}'.format(data_path,
                                               '/'.join(['callCenterDataset',
                                                         x.split('callCenterDataset/')[1]])))
```
